#Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples
-
  -Anomaly detection is the process of identifying rare, suspicious events or data points that deviate significantly from a dataset's normal behavior, often signaling potential problems like fraud, system malfunctions, or cyberattacks. The three main types are Point Anomalies, which are single data points that are significantly different from the rest; Contextual Anomalies, where a data point is an anomaly due to its specific situation or context; and Collective Anomalies, which are a group of data instances that together show an unusual pattern, even if the individual instances appear normal.
Types of Anomalies
1. Point Anomaly
Description: A single data point that is an outlier and doesn't fit the overall pattern or distribution of the dataset.
Example: A credit card transaction for a very high amount that is significantly outside the user's normal spending habits would be a point anomaly, flagging potential fraud. Another example is a sudden 90% CPU usage spike on a server, deviating from its typical 30% usage, as seen in Centreon.
2. Contextual Anomaly
Description: A data point that is unusual not in itself, but in the specific context or situation in which it occurs.
Example: Low water usage might be normal at 3 AM, but it would be considered a contextual anomaly if it occurs during peak usage hours in a residential area. Similarly, a small purchase in a foreign country, occurring at an unusual time for the user, would be anomalous in its context.
3. Collective Anomaly
Description: A collection of data instances that appear normal when viewed individually, but together form an unusual pattern or behavior that signals an anomaly.
Example: A sudden, sustained increase in network traffic over several hours could be a collective anomaly, indicating a potential cyberattack or system overload, even though individual data packets might seem normal.


#Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
-
  -Isolation Forest is a fast, scalable algorithm that detects global anomalies by isolating data points in random decision trees, ideal for high-dimensional data and large datasets. DBSCAN is a density-based clustering algorithm excellent for spatial or geographic data and finding clusters in datasets with non-uniform shapes, though it can be sensitive to parameter settings. Local Outlier Factor (LOF) effectively identifies local anomalies by comparing a point's density to its neighbors, making it useful for finding outliers in dense clusters but less scalable for high-dimensional or massive datasets.
Here's a comparison of the three algorithms:
Isolation Forest
Approach: Builds an ensemble of "isolation trees" where anomalies are isolated in fewer splits (shorter paths from the root) compared to normal data points.
Suitable Use Cases:
High-dimensional data: Handles high-dimensional datasets well.
Large datasets: Its tree-based nature makes it fast and scalable.
Global anomaly detection: Excels at finding data points that are globally different from the rest of the dataset.
Strengths: Fast, scalable, and requires no assumptions about data distribution.
Weaknesses: May struggle with local anomalies that resemble normal points.
DBSCAN
Approach:
A density-based clustering algorithm that groups together points that are closely packed together and marks points in low-density areas as outliers or noise.
Suitable Use Cases:
Spatial or geographic data: Useful for detecting clusters with irregular shapes.
Noisy data: Can handle data with noise.
Clustering: Identifies clusters of varying shapes and sizes without requiring a pre-defined number of clusters.

#Question 3: What are the key components of a Time Series? Explain each with one example?
-
  -The four key components of a time series are: Trend, a long-term upward or downward movement; Seasonal Variation, regular, predictable patterns within a year, such as higher ice cream sales in summer; Cyclical Variation, longer-term, irregular up-and-down movements over several years, like business cycles; and Irregular Variation, unpredictable, random fluctuations caused by unexpected events, such as a spike in sales due to a natural disaster.
Here is a more detailed explanation of each component with an example:
1. Trend
Explanation:
This component represents the long-term progression or general direction of the data over an extended period. It indicates if the data is generally increasing, decreasing, or staying relatively stable.
Example:
The increasing number of mobile payment transactions over several years, driven by consumer adoption of digital payment methods.
2. Seasonal Variation
Explanation:
This refers to periodic, short-term fluctuations that repeat at fixed intervals, typically within a year (e.g., monthly, quarterly).
Example:
Increased sales of ice cream and air conditioners during the warm summer months.
3. Cyclical Variation
Explanation:
These are longer-term oscillations or fluctuations around the trend that do not have a fixed period but are related to economic or other long-term cycles.
Example:
The boom-and-bust phases of a business cycle, where a period of growth is followed by a recession over several years.
4. Irregular (or Random) Variation
Explanation:
This component captures unpredictable, short-term, and random fluctuations or "noise" in the data that are not part of the underlying trend, seasonality, or cyclical patterns.
Example:
A sudden, temporary drop in daily electricity consumption due to an unplanned power outage at a large factory.

#Question 4: Define Stationary in time series. How can you test and transform a
# non-stationary series into a stationary one?

 -
   -A time series is stationary if its statistical properties, like mean and variance, are constant over time and its autocovariance only depends on the time lag, not the specific time point. You can test for stationarity using visual inspection (plotting the data), summary statistics (checking for consistency), or statistical tests like the Dickey-Fuller test or KPSS test. To transform a non-stationary series, you can use differencing (subtracting consecutive values), apply logarithmic transformations to stabilize variance, or detrend the data by removing deterministic trends.  
What is a Stationary Time Series?
A stationary time series has statistical properties that do not change over time. This means:
Constant Mean: The average value of the series remains consistent.
Constant Variance: The spread or volatility of the series does not change over time.
Constant Autocovariance: The relationship (covariance) between an observation and a lagged observation is the same, regardless of the time at which the observations occur.
How to Test for Stationarity
1. Visual Inspection:
Plot the time series to look for obvious trends (upward or downward movement) or seasonality (repeating patterns over specific periods). If there's no clear trend or seasonality, and the data fluctuates around a consistent level, it might be stationary.
2. Summary Statistics:
Divide the data into different segments and compare their means and standard deviations. If these values are roughly the same across segments, the series is likely stationary.
3. Statistical Tests:
Augmented Dickey-Fuller (ADF) Test: This test checks for the presence of a unit root, a characteristic of non-stationary series. If the test indicates no unit root, the series is likely stationary.
Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test: This test checks if a series is stationary around a deterministic trend.
How to Transform a Non-Stationary Series
1. Differencing:
This is a common method for removing trends and stabilizing the mean. You subtract each observation from the next one in the series, creating a new series of differences. For example, Yt - Yt-1.
2. Logarithmic Transformation:
Applying a log transformation can help to stabilize the variance of a time series, making it more consistent over time.
3. De-trending:
This involves fitting a trend (e.g., using linear regression) and then subtracting the trend component from the original data to make the series stationary.
4. Seasonality Adjustment:
If the series has a seasonal pattern, you can remove it by taking differences of observations that are separated by the seasonal period (e.g., differencing observations a year apart to remove a yearly seasonal trend).

#Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
#erms of structure and application.
-
  -AR, MA, ARIMA, SARIMA, and SARIMAX models sequentially build on forecasting techniques: AR uses past values to predict the future, MA models errors using past forecast errors, ARIMA adds differencing to handle non-stationary data and trends, SARIMA introduces seasonal differencing and seasonal components for recurring patterns, and SARIMAX extends SARIMA by incorporating external, non-seasonal variables to improve forecast accuracy.
Here's a breakdown of their structure and applications:
AR (Autoregressive) Model
Structure:
Uses its own past values (lags) to predict future values. Denoted as AR(p), where 'p' is the number of lags.
Application:
Suitable for time series data where the current value is a linear combination of past values, without external factors.
MA (Moving Average) Model
Structure:
Models the current error as a linear combination of past error terms. Denoted as MA(q), where 'q' is the number of past error terms.
Application:
Useful when the forecast error has an autocorrelative structure, where past errors influence current ones.
ARIMA (Autoregressive Integrated Moving Average) Model
Structure:
Combines AR and MA components with a differencing term to make a non-stationary time series stationary. The standard notation is ARIMA(p, d, q), where 'p' is the AR order, 'd' is the differencing order, and 'q' is the MA order.
Application:
Widely used for forecasting non-seasonal, stationary time series data that may exhibit trends or other non-stationary patterns.
SARIMA (Seasonal Autoregressive Integrated Moving Average) Model
Structure:
An extension of ARIMA, SARIMA adds a seasonal component. It includes an additional set of AR, MA, and differencing terms for the seasonal part, specified by (P, D, Q) and the seasonal period 'm'.
Application:
Ideal for time series data with recurring seasonal patterns, such as monthly sales data showing annual cycles or daily electricity usage.
SARIMAX (Seasonal Autoregressive Integrated Moving Average with eXogenous regressors) Model
Structure:
The most comprehensive model, SARIMAX incorporates seasonality, trend, and external variables (exogenous regressors) into the SARIMA framework.
Application:
Used when external factors significantly influence the time series, like predicting hotel occupancy using seasonal patterns along with factors like marketing campaigns or local events.

```python
#● NYC Taxi Fare Data
# AirPassengers Dataset
#Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
#and decompose it into trend, seasonality, and residual components

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 1: Load the AirPassengers dataset
from statsmodels.datasets import get_rdataset
data = get_rdataset("AirPassengers", package="datasets").data

# Convert to datetime and set index
data['Month'] = pd.date_range(start='1949-01-01', periods=len(data), freq='MS')
data.set_index('Month', inplace=True)
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Step 2: Plot the original time series
plt.figure(figsize=(10, 4))
plt.plot(data['Passengers'], label='Monthly Air Passengers')
plt.title("AirPassengers Time Series")
plt.ylabel("Number of Passengers")
plt.xlabel("Date")
plt.legend()
plt.grid(True)
plt.show()

# Step 3: Decompose the time series
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative')

# Plot the decomposed components
decomposition.plot()
plt.tight_layout()
plt.show()
```

```python
#Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
#detect anomalies. Visualize the anomalies on a 2D scatter plot.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Step 1: Simulate a small sample NYC Taxi Fare dataset
np.random.seed(42)
n_samples = 500

# Generate normal trip data
fare_amount = np.random.normal(loc=12, scale=5, size=n_samples)
trip_distance = np.random.normal(loc=2.5, scale=1, size=n_samples)

# Add some anomalies
fare_amount = np.append(fare_amount, [100, 200, 150])  # high fare anomalies
trip_distance = np.append(trip_distance, [0.1, 30, 50])  # abnormal distances

# Create DataFrame
df = pd.DataFrame({'fare_amount': fare_amount, 'trip_distance': trip_distance})

# Step 2: Apply Isolation Forest
clf = IsolationForest(contamination=0.02, random_state=42)
df['anomaly'] = clf.fit_predict(df[['fare_amount', 'trip_distance']])

# Step 3: Visualize
plt.figure(figsize=(10, 6))
normal = df[df['anomaly'] == 1]
anomalies = df[df['anomaly'] == -1]

plt.scatter(normal['trip_distance'], normal['fare_amount'], c='blue', label='Normal')
plt.scatter(anomalies['trip_distance'], anomalies['fare_amount'], c='red', label='Anomaly')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount ($)')
plt.title('NYC Taxi Fare Anomaly Detection using Isolation Forest')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 8: Train a SARIMA model on the monthly airline passengers dataset.
#Forecast the next 12 months and visualize the results.


import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.datasets import get_rdataset

# Step 1: Load the AirPassengers dataset
data = get_rdataset("AirPassengers", package="datasets").data
data['Month'] = pd.date_range(start='1949-01-01', periods=len(data), freq='MS')
data.set_index('Month', inplace=True)
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Step 2: Train SARIMA model
# We'll use SARIMA(p,d,q)(P,D,Q,s)
# This is a quick model, not optimized. AIC/BIC optimization could improve it.
model = SARIMAX(
    data['Passengers'],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()

# Step 3: Forecast the next 12 months
forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data.index[-1] + pd.offsets.MonthBegin(), periods=forecast_steps, freq='MS')
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Step 4: Plot the results
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Passengers'], label='Observed')
plt.plot(forecast_index, forecast_mean, label='Forecast', color='green')
plt.fill_between(
    forecast_index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    color='green',
    alpha=0.3,
    label='95% Confidence Interval'
)
plt.title('AirPassengers Forecast - SARIMA Model')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
#anomalies and visualize them using matplotlib.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 1: Generate synthetic numerical data
np.random.seed(42)
X = np.random.normal(loc=0, scale=1, size=(300, 2))

# Add some anomalies
outliers = np.random.uniform(low=-6, high=6, size=(10, 2))
X = np.vstack([X, outliers])

# Step 2: Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.03)
y_pred = lof.fit_predict(X)

# Add predictions to DataFrame
df = pd.DataFrame(X, columns=["Feature1", "Feature2"])
df["LOF_Label"] = y_pred

# Step 3: Visualize the results
plt.figure(figsize=(10, 6))
normal = df[df["LOF_Label"] == 1]
anomalies = df[df["LOF_Label"] == -1]

plt.scatter(normal["Feature1"], normal["Feature2"], c='blue', label="Normal")
plt.scatter(anomalies["Feature1"], anomalies["Feature2"], c='red', label="Anomaly", marker='x', s=100)
plt.title("Anomaly Detection using Local Outlier Factor (LOF)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
